```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
re_data = pd.read_csv(r'C:\Users\patel d d\Desktop\project csv files\renewable energy dataset (1).csv')
re_data
```

* # Columns:
Year: This is a time-based feature indicating the year. It will help in trend analysis.

Month: A time-based feature showing the month of the year. This allows for identifying seasonal trends.

Sector: Likely denotes the economic or industry sector (e.g., residential, industrial, commercial) involved in renewable energy consumption or production.

* # Energy Sources:

Hydroelectric Power: Energy produced from water resources.

Geothermal Energy: Energy harnessed from the Earth’s internal heat.

Solar Energy: Energy obtained from sunlight.

Wind Energy: Energy harnessed from wind.

Wood Energy: Energy generated from burning wood biomass.

Waste Energy: Energy obtained from the burning or processing of waste materials.

Fuel Ethanol, Excluding Denaturant: Renewable fuel made from biomass materials like corn.

Biomass Losses and Co-products: Could be the losses or by-products of biomass energy production.

Biomass Energy: Energy derived from organic materials (e.g., plant matter, wood).

Total Renewable Energy: The sum of all renewable energy sources.

Renewable Diesel Fuel: Diesel fuel made from renewable resources.

Other Biofuels: Any biofuels not specifically mentioned above (e.g., methanol, butanol).

Conventional Hydroelectric Power: Traditional large-scale hydroelectric power plants.

Biodiesel: A biofuel made from renewable biological sources like vegetable oils or animal fats.

```python
df_year = re_data.groupby('Year').sum().reset_index().drop(columns=['Month', 'Sector'])
df_year
```

Analysis of Renewable Energy

Distribution of Renewable Energy

```python

Renewable_energy = re_data.drop(columns=['Fuel Ethanol, Excluding Denaturant','Renewable Diesel Fuel','Other Biofuels','Conventional Hydroelectric Power','Biodiesel','Year','Month','Sector','Biomass Energy','Total Renewable Energy'])
Renewable_energy_sum = Renewable_energy.sum()
plt.figure(figsize=(14,8))
plt.title('Distribution of Renewable Energy')
plt.xlabel('Energy Type')
plt.xticks(rotation=90)
plt.bar(Renewable_energy_sum.index,Renewable_energy_sum.values)
plt.show()
```

Distrbution of Energy Consumption by Sectors

```python
# Distrbution of Energy Consumption by Sectors
energy_by_sector = re_data.groupby('Sector').sum()
energy_columns = ['Hydroelectric Power', 'Geothermal Energy', 'Solar Energy', 
                  'Wind Energy', 'Wood Energy', 'Waste Energy', 
                  'Fuel Ethanol, Excluding Denaturant', 'Biomass Losses and Co-products',
                  'Renewable Diesel Fuel', 
                  'Other Biofuels', 'Conventional Hydroelectric Power', 'Biodiesel']
energy_by_sector_total = energy_by_sector[energy_columns].sum(axis=1)
plt.figure(figsize=(12,6))
energy_by_sector_total.plot(kind='bar')
plt.title('Sector Energy Consumption')
plt.xlabel('Sectors')
plt.show()
```

# Sector-wise Energy Consumption by Type

```python

sector_energy_sum = re_data.groupby('Sector')[['Hydroelectric Power', 'Geothermal Energy',
                                                'Solar Energy', 'Wind Energy', 'Wood Energy', 'Waste Energy',
                                                  'Fuel Ethanol, Excluding Denaturant', 'Biomass Losses and Co-products',
                                                    'Renewable Diesel Fuel', 'Other Biofuels', 
                                                    'Conventional Hydroelectric Power', 'Biodiesel']].sum()
plt.figure(figsize=(12,6))
sector_energy_sum.plot(kind='bar', stacked=True)
plt.xlabel('Sector')
plt.ylabel('Total Energy Consumption (trillion BTUs)')
plt.title('Total Energy Consumption by Sector')
plt.legend(title='Energy Type', loc='upper left',bbox_to_anchor = (2,1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Columns, Sectors 

cols = ['Hydroelectric Power', 'Geothermal Energy', 'Solar Energy', 'Wind Energy',
       'Waste Energy', 'Fuel Ethanol, Excluding Denaturant', 'Biomass Losses and Co-products',
       'Renewable Diesel Fuel', 'Other Biofuels', 'Conventional Hydroelectric Power',
       'Biodiesel']

sectors = ['Commerical', 'Electric Power', 'Industrial', 'Residential',
           'Transportation']
```

```python
plt.figure(figsize=(12,6))
for energy_type in cols:
    plt.plot(df_year['Year'],df_year[energy_type],label=energy_type)
plt.xlabel("YEAR")
plt.ylabel("TOTAL ENERGY CONSUMPTION(TRILLION BTUS)")
plt.title('Total Energy Consumption by year')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.tight_layout()
plt.show()
```

Wind Energy (Red Line): Rapid growth, becoming the dominant source after 2010, peaking sharply before a drop around 2023.

Conventional Hydroelectric Power (Cyan Line): Historically the largest, fluctuates but stays relatively stable with some decline after 2000.

Solar Energy (Green Line): Steady increase after 2010, showing strong growth.

Fuel Ethanol (Excluding Denaturant) (Pink Line): Strong growth starting in the 2000s, followed by stabilization.

Other Biofuels (Gray Line): Gradual increase from around 2010.

Waste Energy (Brown Line) and Biodiesel (Blue Line): Smaller contributions, showing less variation over time.

 Key Insight:

The chart shows a clear transition to wind and solar energy, overtaking more traditional renewable sources like hydroelectric power as primary contributors to renewable energy consumption. There’s a sharp decline in 2023 across most sources, possibly due to incomplete data or external factors.

# Trend of Renewable Energy Trends Over Time

```python

Trend_Renewable = re_data.groupby('Year')[Renewable_energy.columns].sum()
Trend_Renewable_sum = Trend_Renewable.sum(axis=1)
plt.figure(figsize=(12, 8))
plt.plot(Trend_Renewable_sum.index, Trend_Renewable_sum.values, marker='o', linestyle='-')
plt.xlabel('Year')
plt.ylabel('Total Renewable Energy Consumption')
plt.title('Total Renewable Energy Consumption Over the Years')
plt.grid(True)
plt.show()
```

Steady Growth (1970s–2020): From around 1975 to 2020, there is a consistent upward trend in total renewable energy consumption. 
This reflects an increase in the adoption and utilization of renewable energy sources over the decades.

# Energy per Year

```python
# Energy per Year

fig, axes = plt.subplots(11, 1, figsize=(8, 40))

for ind, col in enumerate(cols):
    plt.subplot(11, 1, ind+1)
    df_top = re_data.groupby(['Year'])[[col]].sum().reset_index()
    df_top = df_top.sort_values(by=(col), ascending=False)

    sns.barplot(x=df_top["Year"], y=df_top[col], palette="Blues")
    plt.yticks(rotation=45, fontsize = 7)
    plt.xticks(rotation=90, fontsize = 7)   
    plt.title(f"Consumption per Year according to {col}", fontsize = 12, color = 'blue')

plt.tight_layout()
```

 # Wind / Solar vs Conventional Hydroelectric Power

```python

df_year.plot(x="Year", y=['Wind Energy', 'Conventional Hydroelectric Power', 'Solar Energy'],
        kind="line", figsize=(10, 6))
plt.title("Wind / Solar vs Conventional Hydroelectric Power", fontsize = 16, color = 'blue')
plt.ylabel('Consumption', rotation=90, fontsize = 10)

plt.grid()
plt.show()
```


Hydroelectric Power was historically the dominant renewable energy source but has not grown as rapidly as newer technologies.

Wind and Solar Energy have experienced rapid growth since the 2000s, driven by advancements in technology and increased investments in clean energy.

The transition to wind and solar suggests a broader shift in renewable energy strategies over the past few decades, with wind energy leading the growth, followed closely by solar energy.


This graph captures the dynamic shift from traditional renewable sources (hydroelectric) to newer forms like wind and solar, which are rapidly increasing in importance.
